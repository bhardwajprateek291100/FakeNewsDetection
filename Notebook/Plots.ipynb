```python
team = ['False','True']
neutral = [38483, 37536]
negative = [675, 41]
positive = [47, 12]
```

```python
x_axis = np.arange(len(team))

# Multi bar Chart

plt.bar(x_axis -0.2, neutral, width=0.2, label = 'neutral')
plt.bar(x_axis +0.2, negative, width=0.2, label = 'negative')
plt.bar(x_axis -0.2, positive, width=0.2, label = 'positive')

# Xticks

plt.xticks(x_axis, team)
plt.xlabel("Label")
plt.ylabel("Count")
plt.title("Figure 3 Sentiment Count of Labels")
# Add legend

plt.legend()

# Display

plt.show()
```
