##**Calculating Accuracy**

```python
def Accuracy(pred1, pred2, pred):
  cnt = 0
  for i in range(len(pred1)):
    val = ((1.7*pred1[i] + pred2[i])/2.7)
    if(val > 0.5):
      val = 1
    else:
      val = 0

    if val == pred[i]:
      cnt += 1

  return (cnt / len(pred1))
```

```python
Accuracy(model1_train_pred['predictions'].values, model2_train_pred['predictions'].values, train_data['label'].values)
```

```python
Accuracy(model1_test_pred['predictions'].values, model2_test_pred['predictions'].values, test_data['label'].values)
```
